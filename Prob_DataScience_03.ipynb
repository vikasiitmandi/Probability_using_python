# Problem 1 on Central limit theorem

random_integer_data.txt file contains 4000 values. Take 10,000 samples of size of 50, compute means for each sample and make a histogram to verify CLT. 

```python
with open('/content/random_integer_data') as f:
    data=[int(line) for line in f.readlines()]
print(data)
```

```python
lst=[]
for i in range(10000):
  sample=random.choices(data,k=50)
  sample_mean=np.mean(sample)
  lst.append(sample_mean)
len(lst)
```

```python
plt.hist(lst,bins=100)
plt.axvline(np.mean(lst),color='r')
plt.axvline(np.mean(data),color='g')
plt.axvline(np.std(lst),color='y')
plt.axvline(np.std(data)/np.sqrt(50),color='m')
plt.show()
```

# Problem 2 on Weak law of large number 

random_integer_data.txt file is given to you contains 4000 values. From the given txt, take 50 samples for n=10,11,...,1000. Find the mean of samples for each n, compute overall means and plot it to verify the weak law of large numbers.

```python
arr=[]
for i in range(10,1000):
  temp=[]
  for j in range(i):
    sample=random.choices(data,k=50)
    sample_mean=np.mean(sample)
    temp.append(sample_mean)
  arr.append(np.mean(temp))
plt.plot(arr)
plt.axhline(np.mean(data),color='r')
```

# Problem 3 on  Bayes theorem

There are two identical urns containing respectively 6 black and 4 red balls, 2 black and 2 red balls. An urn is chosen at random and a ball is drawn from it. 

(i) Find the probability that the ball is black with the help of simulation (1000 times). Plot the probability of simulated result and observe the difference with theoratical value. [Hint: Find P(Black|Urn1), P(Black|Urn2) to find P(Black).]

(ii) If the ball is black, what is the probability that it is from the first urn? Repeat the process as in (i). 

```python
""" Total number of balls in Urn 1 
Total number of balls in Urn 1 

  =  6 black + 4 red 

  =  10 balls

Probability of getting black ball from Urn 1 

P(B/U1)  =  6/10

Total number of balls in Urn 2

  =  2 black + 2 red 

  =  4 balls

Probability of getting black ball from Urn 2 

P(B/U2)  =  2/4

P(U1)  =  1/2 and P(U2)  =  1/2

(i) find the probability that the ball is black

P(B)  =  P(U1) ⋅ (P(B/U1) +  P(U2) ⋅ (P(B/U2)

  =  (1/2)  ⋅ (6/10) +  (1/2) ⋅ (2/4)

  =  3/10 + 1/4

 =  11/20
 

 (ii) if the ball is black what is the probability that it is from the first urn?

We need to find the probability that the selected black ball is from first urn.
 =  P(U1) ⋅ (P(B/U1) / [P(U1) ⋅ (P(B/U1) +  P(U2) ⋅ (P(B/U2)]

  =   (1/2)  ⋅ (6/10) / [(1/2)  ⋅ (6/10) +  (1/2) ⋅ (2/4)]

  =  (6/20) / (11/20)

  =  6/11
 
 """
```

```python
urn1=['black','black','black','black','black','black','red','red','red','red','red']
urn2=['black','black','red','red']
pb_u1=[]
for i in range(1000):
  x=[]
  for i in range(1000):
    random=np.random.choice(urn1)
    if random=='black':
      x.append(random)
  p1=x.count('black')/1000
  pb_u1.append(p1)
plt.plot(pb_u1)
plt.axhline(0.6,color='r')
plt.axhline(np.mean(pb_u1),color='y')
plt.show()
```

```python
pb_u2=[]
for i in range(1000):
  y=[]
  for i in range(1000):
    random=np.random.choice(urn2)
    if random=='black':
      y.append(random)
  p2=y.count('black')/1000
  pb_u2.append(p2)
plt.plot(pb_u2)
plt.axhline(0.5,color='r')
plt.axhline(np.mean(pb_u2),color='y')
plt.show()
```

```python
# (i) find the probability that the ball is black
# P(B)  =  P(U1) ⋅ (P(B/U1) +  P(U2) ⋅ (P(B/U2)
# P(U1)  =  1/2 and P(U2)  =  1/2
pu1=0.5
pb2=0.5
PB = np.array(pb_u1)*0.5+np.array(pb_u2)*0.5
plt.plot(list(PB))
plt.axhline(0.55,color='r')
plt.axhline(np.mean(PB),color='y')
plt.show()
```

```python
# (ii) if the ball is black what is the probability that it is from the first urn?

# We need to find the probability that the selected black ball is from first urn.

 #=  P(U1) ⋅ (P(B/U1) / [P(U1) ⋅ (P(B/U1) +  P(U2) ⋅ (P(B/U2)] 

PB_urn1 = 0.5*np.array(pb_u1)/np.array(PB)
plt.plot(PB_urn1)
plt.axhline(6/11,color='r')
plt.axhline(np.mean(PB_urn1),color='y')
```
