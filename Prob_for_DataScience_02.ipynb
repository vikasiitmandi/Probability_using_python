Suppose you have an unfair coin that has a probability of landing heads of p = 0.6, and a
probability of landing tails of 1 âˆ’ p. Write a program to calculate the 

following probabilities:

(a) Probability of getting exactly 2 heads in 5 tosses of the coin.

(b) Probability of getting at least 3 heads in 5 tosses of the coin.

(c) Probability of getting at most 2 heads in 5 tosses of the coin.


```python
import numpy as np
import matplotlib.pyplot as plt

list_prob=[]
for i in range(1000):
  total=[]
  for i in range(1000):
    sample= np.random.choice([+1,-1],5, p=[0.6,0.4])
    count=list(sample).count(+1)
    if count==2:
      total.append(count)
  p=total.count(2)/1000
  list_prob.append(p)
plt.plot(list_prob)
plt.axhline(np.mean(list_prob),color='r')
# calculate by using binomial theorem is 0.234
plt.axhline(0.234,color='g')
```

```python
import numpy as np
import matplotlib.pyplot as plt

list_prob_atleast=[]
for i in range(1000):
  total_atleast=[]
  for i in range(1000):
    sample_atleast= np.random.choice([+1,-1],5, p=[0.6,0.4])
    count=list(sample_atleast).count(+1)
    if count>2:
      total_atleast.append(count)
  p=(total_atleast.count(3)+total_atleast.count(4)+total_atleast.count(5))/1000
  list_prob_atleast.append(p)
plt.plot(list_prob_atleast)
plt.axhline(np.mean(list_prob_atleast),color='r')
```

# PROBLEM 2

As we conduct a single coin toss, the probability of getting heads is 0.5.
By increasing the number of tosses and examining the outcomes, we can calculate 
the observed frequency of heads. As the number of tosses becomes larger, the 
observed frequency of heads tends to converge to 0.5 ?

Note :
This convergence is supported by the Law of Large Numbers, which states that as the number of independent samples increases, the average of those samples approaches the expected value.

```python
import random
ran=[]
for i in range(10,1000):
  temp=[]
  for i in range(i):
    a=random.randint(0,1)
    temp.append(a)
  prob=(temp.count(0)/len(temp))
  ran.append(prob)
plt.plot(ran)
plt.axhline(np.mean(ran),color='r')
```


The probability of getting a specific number on a fair six-sided die is 1/6. With more tosses and examining the outcomes, the observed frequency of that specific number tends to converge to 1/6 as the number of tosses increases ?

Note : This convergence is supported by the Law of Large Numbers, which states that as the number of independent samples increases, the average of those samples approaches the expected value.

```python
ran=[]
for i in range(10,1000):
  temp=[]
  for i in range(i):
    a=random.randint(1,6)
    temp.append(a)
  prob=(temp.count(1)/len(temp))
  ran.append(prob)
plt.plot(ran)
plt.axhline(np.mean(ran),color='r')
```
